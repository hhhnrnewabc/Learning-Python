```python
# coding=utf-8
import urllib.request
import time
import datetime
from lxml import etree
from pandas import DataFrame, read_csv
```

```python
url = 'http://zh.wikipedia.org/zh-tw/%E6%97%A5%E6%9C%AC%E5%8B%95%E7%95%AB%E5%88%97%E8%A1%A8_%282010%E5%B9%B4%29'
# url = 'http://zh.wikipedia.org/zh-tw/%E6%97%A5%E6%9C%AC%E5%8B%95%E7%95%AB%E5%88%97%E8%A1%A8_%282000%E5%B9%B4%E4%BB%A3%29'
```

```python
response = urllib.request.urlopen(url)
```

```python
page = etree.HTML(response.read())
```

```python
path_tables = u"//div/div/div[@class='mw-content-ltr']/table[@class='wikitable']"
path_table_head = u".//tr/th/text()"
path_table_tr = u".//tr[position()>=2]"
path_table_row1 = u".//td[1]/text()"
path_table_row2 = u".//td[2]/a/text()"
path_table_row3_en= u".//td[3]/text()"
path_table_row3_jp= u".//td[3]/span/text()"
path_table_row4_a = u".//td[4]/a/text()"
path_table_row4_text = u".//td[4]/text()"
path_table_row5 = u".//td[5]/text()"
```

```python
tables = page.xpath(path_tables)
tables
```

```python
tables[0].xpath(path_table_head)
```

```python
tables[0].xpath(path_table_tr)[0].xpath(path_table_row1)
```

```python
def get_first_data_from_list(l):
    if len(l)>0:
        return l[0]
    else:
        return ''
```

```python
output_tables = []
for idx, table in enumerate(tables):
    # print(table.xpath(path_table_head))
    output_data = []
    #output_data.append([table.xpath(path_table_head)])
    for row in table.xpath(path_table_tr):
        r1 = row.xpath(path_table_row1)
        r2 = row.xpath(path_table_row2)
        r3 = row.xpath(path_table_row3_en) if row.xpath(path_table_row3_en) else row.xpath(path_table_row3_jp)      
        r4 = row.xpath(path_table_row4_a) if row.xpath(path_table_row4_a) else row.xpath(path_table_row4_text)
        r5 = row.xpath(path_table_row5)
        
        r1 = get_first_data_from_list(r1)
        r2 = get_first_data_from_list(r2)
        r3 = get_first_data_from_list(r3)
        r4 = get_first_data_from_list(r4)
        r5 = get_first_data_from_list(r5)
        
        output_data.append([r1, r2, r3, r4, r5])

    output_table = DataFrame(data=output_data, columns=table.xpath(path_table_head))
    output_tables.append(output_table)
    output_table.to_csv('Anime_2010_%02d.csv' % idx,index=False,header=False)
```
