```python
import lxml.etree
import datetime
```

```python
start_week = 127
end_week = 761+1

xpath_url = """//table/tr[1]/td[2]/strong[3]/text()"""
```

```python
# General syntax to import specific functions in a library: 
##from (library) import (specific library function)
from pandas import DataFrame, read_csv, to_datetime

# General syntax to import a library but no functions: 
##import (library) as (give the library a nickname/alias)
import matplotlib.pyplot as plt
import pandas as pd #only needed to determine version number
import sys #only needed to determine Python version number

plt.rcParams['font.sans-serif'] = ['LiHei Pro']

# Enable inline plotting
%matplotlib inline
```

```python
oil_data = []
time = datetime.datetime.strptime ("2002/6/2", "%Y/%m/%d")
for filename in range(start_week, end_week):
    with open('./%s.html' % filename, encoding='big5')as data:
        page = lxml.etree.HTML(data.read())
        oil_raw = page.xpath(xpath_url)
        try:
            oil = oil_raw[0].split('：')[1]
            oil_data.append([filename, oil, time])
        except IndexError :
            print("IndexError  %3d" % filename)
        time += datetime.timedelta(days=7)
```

```python
oil_data_ps = DataFrame(data=oil_data,
                        columns=['週次', '週平均價', '時間'])

#oil_data_ps = DataFrame(data=oil_data,
#                        columns=['週次', '週平均價', '時間'],
#                        index=oil_data_ps['時間'])

oil_data_ps.to_csv('oil.csv',index=False,header=True)
```

```python
plt.figure()

oil_data_ps['週平均價'].plot(figsize=(18, 6),style='d-', label='油價')

MaxValue = oil_data_ps['週平均價'].max()
MaxWeek = oil_data_ps['週次'][oil_data_ps['週平均價'] == oil_data_ps['週平均價'].max()].values
#MaxDate = oil_data_ps['週次'][oil_data_ps['週平均價'] == oil_data_ps['週平均價'].max()].index.values

MinValue = oil_data_ps['週平均價'].min()
MinWeek = oil_data_ps['週次'][oil_data_ps['週平均價'] == oil_data_ps['週平均價'].min()].values

Text = str(MaxValue) + " - " + str(MaxWeek) + " 週"

plt.annotate(Text, xy=(1, MaxValue), xytext=(14, 0), 
                 xycoords=('axes fraction', 'data'), textcoords='offset points')


Text = str(MinValue) + " - " + str(MinWeek) + " 週"

plt.annotate(Text, xy=(1, MinValue), xytext=(14, 0), 
                 xycoords=('axes fraction', 'data'), textcoords='offset points')
plt.legend(loc='best')
```
