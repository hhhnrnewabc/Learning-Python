```python
white_ip = ['140.134.247.177',]
```

```python
base_path = '/Users/swim/Desktop/nginx/'
log_list = ['access.log.4', 'access.log.3']
```

```python
ip_dict = dict()
```

```python
count = 0
for f in log_list:
    file_name = base_path + f
    with open(file_name) as log:
        for data in log:
            # print(data)
            mdict = log_re.match(data).groupdict()
            http_code = mdict.get('HTTP_code', None)
            ip = mdict.get('ip', None)
            if ip == '61.191.62.7':
                    print(data, end='')
            if (ip not in white_ip ) and (http_code != '200' and http_code != '302'):
                #print(data, end='')
                ip_dict[ip] = ip_dict.get(ip, 0) + 1 
                count += 1
                
print("Line Number:%4d" % count)
```

```python
max_ip_dict = {"ip":[], "times":0}
for ip in ip_dict:
    if ip_dict[ip] > max_ip_dict['times']:
        max_ip_dict['ip'] = [ip]
        max_ip_dict['times'] = ip_dict[ip]
    elif ip_dict[ip] == max_ip_dict['times']:
        max_ip_dict['ip'] += [ip]
max_ip_dict
```

```python
from IPython.core.display import Image
import graph_tool.all as gt
```

```python
g = gt.Graph()
vprop_vstring = g.new_vertex_property("string")
vprop_times = g.new_vertex_property("int")  
```

```python
for ip in ip_dict:
    v = g.add_vertex()
    vprop_vstring[v] = ip
    vprop_times[v] = ip_dict[ip]
```

```python
g.vp['string'] = vprop_vstring
g.vp['int'] = vprop_times
```

```python
g.list_properties()
```

```python
v_size_p = gt.prop_to_size(
    vprop_times,
    10,
    28, )
```

```python
gt.graph_draw(g, vertex_text=vprop_vstring, vertex_font_size=v_size_p,
              vertex_fill_color = v_size_p,
              output_size=(2500, 2000), output="ip.png")
```
